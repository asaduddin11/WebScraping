```
json_data = json.loads(response.text)
all_title = []
all_links = []
inventors = []
description = []
pub_dates = []
```

```
for items in json_data['items']:
  patent_response = requests.get(items['link'])
  soup = BeautifulSoup(patent_response.text, "html5lib")
  time.sleep(1)
  try:
    all_title.append([p.getText() for p in soup.find_all("h1")][0])
  except: 
    all_title.append("")
  try:
    description.append([d.getText() for d in soup.find_all('div', itemprop = "content")][0])
  except:
    description.append("")
  try:
    all_links.append(items['link'])
  except: 
    all_links.append("")
  try:
    pub_dates.append(soup.find('td',itemprop='publicationDate').text)
  except:
    pub_dates.append("")
  try:
    inventors.append([x.getText() for x in soup.find_all('dd',itemprop='inventor')])
  except:
    inventors.append("")
  

patent_link_data = pd.DataFrame({'Title': all_title, 'Description': description,
                         'Link': all_links, 'Publishing Date': pub_dates, 'Inventors': inventors})

writer = pd.ExcelWriter('patents2.xlsx')
patent_link_data.to_excel(writer, 'Patents')
writer.save()
```
